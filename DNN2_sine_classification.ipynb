```python
import matplotlib.pyplot as plt
import numpy as np
from keras.layers import *
from keras.models import *
from keras.utils import *
from collections import Counter
%matplotlib inline
```

```python
x = np.random.uniform(-np.pi, np.pi, 500)
y = np.random.uniform(-1, 1, 500)
# plt.scatter(x, y)
# plt.show() # matplotlib inline을 하면 이거 없어도 된데!

X = np.asarray([[x[i], y[i]] for i in range(500)])
# print(X)
print(X.shape)
# plt.scatter(X[:,0], X[:,1])
```

```python
sine = np.sin(x)
plt.scatter(x, sine)
plt.scatter(X[:,0], X[:,1])
```

```python
Y = sine < X[:,1]
plt.scatter(x, sine)
# plt.scatter(X[:,0][Y==0], X[:,1][Y==0])
plt.scatter(X[:,0][Y==False], X[:,1][Y==False]) # 이렇게 하는 게 맞는거 같아...
# plt.scatter(X[:,0][Y==1], X[:,1][Y==1])
plt.scatter(X[:,0][Y==True], X[:,1][Y==True]) # 이렇게 하는 게 맞는거 같아...
```

```python
Y = to_categorical(Y)
Y
```

```python
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(2,)))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy', optimizer='sgd')

model.fit(X, Y, epochs=200)

```

```python
score = model.evaluate(X, Y)
print(score)
```

```python
pred_x = np.vstack(([-2, 0], [2,0]))
print(pred_x)

pred = model.predict(pred_x)
print(pred)
```
